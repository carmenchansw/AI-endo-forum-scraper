```python
from concurrent.futures import ThreadPoolExecutor
from playwright.sync_api import sync_playwright
from bs4 import BeautifulSoup

url = 'https://endometriosis.net/forums/advice-for-potential-endometriosis'

def scrape_with_playwright(target_url):
    # This runs in a separate thread where Windows/Jupyter cannot interfere
    with sync_playwright() as p:
        print("Launching browser in background thread...")
        browser = p.chromium.launch(headless=True)
        page = browser.new_page()
        
        print("Navigating to page...")
        page.goto(target_url, wait_until="networkidle")
        
        # Pull HTML and extract the text cleanly
        html = page.content()
        soup = BeautifulSoup(html, 'html.parser')
        page_title = soup.title.text if soup.title else "No title found"
        
        browser.close()
        return page_title

# Force the execution outside of Jupyter's loop space
with ThreadPoolExecutor(max_workers=1) as executor:
    future = executor.submit(scrape_with_playwright, url)
    result = future.result()  # Grab the returned title

print(f"\nSuccess! Page Title extracted: {result}")
```
